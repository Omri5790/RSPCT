```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# קריאת הנתונים מקובץ ה-CSV המעודכן עם הקלאסטרים
df = pd.read_csv('ORIGINS_shot_data.csv')

# הוספת הקלאסטרים לנתונים לפי הקוד הקודם
features = df[['Distance', 'Cumulative_Distance', 'ShotDistance', 'ShotAngle']]
optimal_k = 3  # בחירת מספר הקבוצות לפי שיטת המרפק מהגרף הקודם
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(features)

# ויזואליזציה 1: מרחק הזריקה מהסל לאורך הזריקות
plt.figure(figsize=(12, 6))
for cluster in df['Cluster'].unique():
    subset = df[df['Cluster'] == cluster]
    plt.plot(subset.index, subset['ShotDistance'], label=f'Cluster {cluster}')
plt.xlabel('Shot Index')
plt.ylabel('Shot Distance')
plt.title('Shot Distance over Time by Cluster')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()

# ויזואליזציה 2: זווית הזריקה לאורך הזריקות
plt.figure(figsize=(12, 6))
for cluster in df['Cluster'].unique():
    subset = df[df['Cluster'] == cluster]
    plt.plot(subset.index, subset['ShotAngle'], label=f'Cluster {cluster}')
plt.xlabel('Shot Index')
plt.ylabel('Shot Angle')
plt.title('Shot Angle over Time by Cluster')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()

# ויזואליזציה 3: מרחק בין זריקות לאורך הזריקות
plt.figure(figsize=(12, 6))
for cluster in df['Cluster'].unique():
    subset = df[df['Cluster'] == cluster]
    plt.plot(subset.index, subset['Distance'], label=f'Cluster {cluster}')
plt.xlabel('Shot Index')
plt.ylabel('Distance Between Shots')
plt.title('Distance Between Shots over Time by Cluster')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# קריאת הנתונים מקובץ ה-CSV המעודכן עם מרחקים
df = pd.read_csv('ORIGINS_shot_data.csv')

# בחירת התכונות לנירמול: מרחק, מרחק מצטבר, ShotDistance, ShotAngle (בהנחה שהן לא מנורמלות)
features_to_normalize = ['Distance', 'Cumulative_Distance', 'ShotDistance', 'ShotAngle']

# נירמול התכונות שנבחרו
scaler = StandardScaler()
df[features_to_normalize] = scaler.fit_transform(df[features_to_normalize])

# שימוש בשיטת המרפק למציאת מספר הקבוצות האופטימלי
features = df[features_to_normalize]
sse = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features)
    sse.append(kmeans.inertia_)

# גרף של שיטת המרפק
plt.figure(figsize=(8, 5))
plt.plot(k_values, sse, '-o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method for Optimal k')
plt.xticks(k_values)
plt.grid(True)
plt.show()

# בחירת מספר הקבוצות מהגרף וביצוע קלאסטרינג עם K-means
optimal_k = 6 # יש לבחור את הערך לפי הגרף של שיטת המרפק
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Temp_Cluster'] = kmeans.fit_predict(features)

# תיקון הקלאסטרינג לפי ClusterID
cluster_mapping = {}

# התאמת הקלאסטרים כך שכל מקבץ זריקות עם אותו ClusterID יהיה באותו קלאסטר
for cluster_id in df['ClusterID'].unique():
    temp_cluster = df.loc[df['ClusterID'] == cluster_id, 'Temp_Cluster'].mode()[0]
    cluster_mapping[cluster_id] = temp_cluster

# עדכון התוויות בהתאם
df['Cluster'] = df['ClusterID'].map(cluster_mapping)

# ויזואליזציה 1: מרחק הזריקה מהסל לאורך הזריקות - גרף לכל קלאסטר
for cluster in df['Cluster'].unique():
    plt.figure(figsize=(12, 6))
    subset = df[df['Cluster'] == cluster]
    plt.plot(subset.index, subset['ShotDistance'], label=f'Cluster {cluster}')
    plt.xlabel('Shot Index')
    plt.ylabel('Shot Distance')
    plt.title(f'Shot Distance over Time for Cluster {cluster}')
    plt.grid(True)
    plt.legend()
    plt.show()

# ויזואליזציה 2: זווית הזריקה לאורך הזריקות - גרף לכל קלאסטר
for cluster in df['Cluster'].unique():
    plt.figure(figsize=(12, 6))
    subset = df[df['Cluster'] == cluster]
    plt.plot(subset.index, subset['ShotAngle'], label=f'Cluster {cluster}')
    plt.xlabel('Shot Index')
    plt.ylabel('Shot Angle')
    plt.title(f'Shot Angle over Time for Cluster {cluster}')
    plt.grid(True)
    plt.legend()
    plt.show()

# ויזואליזציה 3: מרחק בין זריקות לאורך הזריקות - גרף לכל קלאסטר
for cluster in df['Cluster'].unique():
    plt.figure(figsize=(12, 6))
    subset = df[df['Cluster'] == cluster]
    plt.plot(subset.index, subset['Distance'], label=f'Cluster {cluster}')
    plt.xlabel('Shot Index')
    plt.ylabel('Distance Between Shots')
    plt.title(f'Distance Between Shots over Time for Cluster {cluster}')
    plt.grid(True)
    plt.legend()
    plt.show()
```
