```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt

# שקלול המיקום לעומת הזמן (לדוגמה, משקל פי 2 למיקום)
df['Weighted_Angle'] = df['ShotAngle'] * 1
df['Weighted_Distance'] = df['ShotDistance'] * 1
df['Weighted_Time'] = df['TimeInSeconds'] * 1

# יצירת דאטה משוקלל לקלאסטרינג
weighted_data = df[['Weighted_Angle', 'Weighted_Distance', 'Weighted_Time','OriginX','OriginY', 'Position', '3P-Range']]

# נורמליזציה של הנתונים כדי לשפר את ביצועי הקלאסטרינג
scaler = StandardScaler()
scaled_data = scaler.fit_transform(weighted_data)

# טווחי הערכים ל-min_samples וערך קבוע עבור eps
min_samples_values = range(2, 10)  # טווח ערכים עבור min_samples
eps = 0.8  # ערך קבוע עבור eps

# שמירת תוצאות
results = []

# הרצת DBSCAN עבור כל ערך של min_samples
for min_samples in min_samples_values:
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    labels = dbscan.fit_predict(scaled_data)
    
    # בדיקה אם יש יותר מקלאסטר אחד תקין
    if len(set(labels)) > 1 and len(set(labels)) - (1 if -1 in labels else 0) > 1:
        silhouette = silhouette_score(scaled_data, labels)
        davies_bouldin = davies_bouldin_score(scaled_data, labels)
    else:
        silhouette = -1  # ערך שמצביע על קלאסטרינג לא תקין
        davies_bouldin = float('inf')  # אינסוף מצביע על קלאסטרינג לא תקין

    # שמירת התוצאות
    results.append({
        'min_samples': min_samples,
        'silhouette': silhouette,
        'davies_bouldin': davies_bouldin
    })

# המרת התוצאות לדאטה פריים להצגה בטבלה
results_df = pd.DataFrame(results)

# הצגת הטבלה
print(results_df)

# ויזואליזציה של Silhouette Score מול min_samples
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(results_df['min_samples'], results_df['silhouette'], marker='o', linestyle='-')
plt.title('Silhouette Score vs. Min Samples')
plt.xlabel('Min Samples')
plt.ylabel('Silhouette Score')
plt.grid(True)

# ויזואליזציה של Davies-Bouldin Index מול min_samples
plt.subplot(1, 2, 2)
plt.plot(results_df['min_samples'], results_df['davies_bouldin'], marker='o', linestyle='-')
plt.title('Davies-Bouldin Index vs. Min Samples')
plt.xlabel('Min Samples')
plt.ylabel('Davies-Bouldin Index')
plt.grid(True)

plt.tight_layout()
plt.show()
```
